```python
import PyPDF2

def copiar_pdf(input_pdf_path, output_pdf_path):
    # Abre el archivo PDF de entrada y el archivo de salida
    with open(input_pdf_path, 'rb') as archivo_entrada, open(output_pdf_path, 'wb') as archivo_salida:
        # Crea un objeto PDFWriter
        escritor_pdf = PyPDF2.PdfWriter()

        # Crea un objeto PDFReader para el archivo de entrada
        lector_pdf = PyPDF2.PdfReader(archivo_entrada)

        # Copia todas las páginas del lector al escritor
        for num_pagina in range(len(lector_pdf.pages)):
            pagina = lector_pdf.pages[num_pagina]
            escritor_pdf.add_page(pagina)

        # Guarda el PDF copiado en el archivo de salida
        escritor_pdf.write(archivo_salida)

# Ruta del archivo PDF de entrada
pdf_existente = 'plantilla.pdf'

# Ruta de salida del nuevo PDF generado
pdf_generado = 'nuevo_pdf_generado.pdf'

# Llama a la función para copiar el PDF existente
copiar_pdf(pdf_existente, pdf_generado)
```
